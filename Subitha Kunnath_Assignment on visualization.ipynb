```python
import pandas as pnd
df = pnd.read_csv('data.csv')
x = df["Calories"].mean()
df["Calories"].fillna(x, inplace = True) # to replace empty values with mean of the specific column
print(df.to_string()) #Line no:18,28
```

```python
import matplotlib.pyplot as plt
import csv
  
x = []
y = []
  
with open('data.csv','r') as csvfile:
    plots = csv.reader(csvfile, delimiter = ',')
      
    for row in plots:
        x.append(row[2])
        y.append(row[4])
  
plt.bar(x, y, color = 'g', width = 0.72, label = "Calories")
plt.xlabel('Pulse')
plt.ylabel('Calories')
plt.title('Pulse- Calorie relation')
plt.legend()
plt.show()
```

```python
x = []
y = []
  
with open('data.csv','r') as csvfile:
    lines = csv.reader(csvfile, delimiter=',')
    for row in lines:
        x.append(row[1])
        y.append(row[4])
  
plt.plot(x, y, color = 'r', linestyle = 'dashed',
         marker = 'p',label = "Calories in different Dates")
  
plt.xticks(rotation = 90)
plt.xlabel('Dates')
plt.ylabel('Calories')
plt.title('Calorie in Dates ', fontsize = 20)
plt.grid()
plt.legend()
plt.show()
```

```python
Names = []
Values = []
  
with open('data.csv','r') as csvfile:
    lines = csv.reader(csvfile, delimiter=',')
    for row in lines:
        Names.append(row[0])
        Values.append(row[4])
  
plt.scatter(Names, Values, color = 'b',s = 25)
plt.xticks(rotation = 25)
plt.xlabel('Duration')
plt.ylabel('Calories')
plt.title('Calories burnt in specific Duration', fontsize = 10)
  
plt.show()
```
